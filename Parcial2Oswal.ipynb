```python
# Solución Pregunta 1
import matplotlib.pyplot as plt

meses = ["Ene", "Feb", "Mar", "Abr", "May", "Jun", "Jul", "Ago", "Sep", "Oct", "Nov", "Dic"]
temperaturas = [22, 24, 27, 30, 32, 35, 34, 33, 30, 27, 24, 22]

plt.figure(figsize=(8, 6))
plt.plot(meses, temperaturas, color="red", marker="o")
plt.title("Temperatura media mensual")
plt.xlabel("Mes")
plt.ylabel("Temperatura (°C)")
plt.grid(True)
plt.show()
```

```python
# Solución Pregunta 2
import seaborn as sns
import pandas as pd
import numpy as np

# Datos sintéticos
np.random.seed(0)
df2 = pd.DataFrame({
    'Edad': np.random.randint(18, 60, 20),
    'Ingreso': np.random.randint(1000, 5000, 20),
    'Genero': np.random.choice(['Masculino', 'Femenino'], 20)
})

sns.set_style("whitegrid")
plt.figure(figsize=(8, 7))
sns.scatterplot(data=df2, x="Edad", y="Ingreso", hue="Genero", s=80)
plt.title("Relación entre edad e ingreso mensual")
plt.show()
```

```python
# Solución Pregunta 3
datos = np.random.normal(50, 10, 200)

plt.figure(figsize=(6, 8))
plt.hist(datos, bins=15, color="blue", alpha=0.6, edgecolor="black")
plt.title("Distribución de datos (Normal)")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
```

```python

# Solución Pregunta 4
df4 = pd.DataFrame({
    'Altura': np.concatenate([
        np.random.normal(1.3, 0.1, 30),
        np.random.normal(1.7, 0.1, 30),
        np.random.normal(1.65, 0.08, 30)
    ]),
    'Grupo': ['Niños']*30 + ['Adultos']*30 + ['Adultos Mayores']*30
})

plt.figure(figsize=(10, 5))
sns.boxplot(data=df4, x="Grupo", y="Altura")
plt.title("Comparación de alturas por grupo")
plt.show()
```

```python

# Solución Pregunta 5
meses = list(range(1, 13))
regiones = ['Norte', 'Centro', 'Sur']

df1 = pd.DataFrame({
    'Mes': np.tile(meses, len(regiones)),
    'Region': np.repeat(regiones, len(meses)),
    'Ventas': np.random.randint(1000, 5000, len(meses)*len(regiones))
})

# Ventas totales por mes
ventastotal = df1.groupby('Mes')['Ventas'].sum().reset_index()
ventastotal['Promediomovil'] = ventastotal['Ventas'].rolling(window=3, min_periods=1).mean()

plt.figure(figsize=(9, 5))
sns.barplot(data=ventastotal, x="Mes", y="Ventas", color="skyblue", label="Ventas totales")
plt.plot(ventastotal['Mes'], ventastotal['Promediomovil'], color="red", marker="o", label="Promedio móvil (3 meses)")
plt.title("Ventas totales y promedio móvil")
plt.xlabel("Mes")
plt.ylabel("Ventas")
plt.legend()
plt.show()
```
